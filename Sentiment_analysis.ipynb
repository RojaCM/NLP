```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow_datasets as tfds
```

```python
dataset=tfds.load('amazon_us_reviews/Mobile_Electronics_v1_00',split='train',shuffle_files=True)
```

```python
# Create empty lists for the desired columns
review_body_list = []
star_rating_list = []

# Iterate through the dataset and extract values
for example in dataset:
    review_body = example['data']['review_body'].numpy().decode('utf-8')
    star_rating = example['data']['star_rating'].numpy()

    review_body_list.append(review_body)
    star_rating_list.append(star_rating)

# Create a dataframe from the lists
df = pd.DataFrame({'review_body': review_body_list, 'star_rating': star_rating_list})
```

```python
data=df.sample(frac=0.3,random_state=42).reset_index(drop=True)
```

```python
data['sentiment']=data['star_rating'].apply(lambda x:'positive' if x>=3 else 'negative')
```

```python
data['sentiment']=data['sentiment'].map({'positive':1,'negative':0})
```

```python
data.rename(columns={'review_body':'review'},inplace=True)
```

```python
data['review']=data['review'].apply(str.lower)
```

```python
#balancing the datasets

positive=data[data['sentiment']==1].sample(n=7425,random_state=42)
```

```python
negative=data[data['sentiment']==0]
```

```python
train_data=pd.concat([positive,negative]).sample(frac=1).reset_index(drop=True)
```

```python
train_data
```

```python
#sample reviews and labels
reviews=train_data['review'].tolist()
labels=train_data['sentiment'].tolist()
```

```python
#split the dataset into train,test and validation

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(reviews,labels,test_size=0.2,random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
#load the tokenizer
!pip install transformers
from transformers import DistilBertTokenizerFast
model_name='distilbert-base-uncased-finetuned-sst-2-english'
tokenizer= DistilBertTokenizerFast.from_pretrained(model_name)
```

```python
#sample tokenization
print('sentence',x_train[0])
tokenizer(x_train[0],truncation=True,padding=True,max_length=256,return_tensors="tf")
```

#PERFORM TOKENIZATION

```python
MAX_LENGTH=256
train_encodings=tokenizer(x_train,truncation=True,padding=True,max_length=256,return_tensors="tf")
```

```python
test_encodings=tokenizer(x_test,truncation=True,padding=True,max_length=256,return_tensors="tf")
```

```python
import tensorflow as tf
```

```python
#convert the tokenizer to TF object
train_dataset=tf.data.Dataset.from_tensor_slices((dict(train_encodings),y_train))
```

```python
test_dataset=tf.data.Dataset.from_tensor_slices((dict(test_encodings),y_test))
```

```python
import tensorflow as tf
```

```python
from transformers import TFDistilBertForSequenceClassification
```

```python
model=TFDistilBertForSequenceClassification.from_pretrained(pretrained_model_name_or_path=model_name,num_labels=2)
```

```python
optimizer=tf.keras.optimizers.Adam(learning_rate=5e-5)
```

```python
loss_func=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
model.compile(optimizer=optimizer,loss=loss_func,metrics=['accuracy'])
```

```python
model.summary()
```

```python
batch_size=12
epochs=2
```

```python
model.fit(train_dataset.shuffle(len(x_train)).batch(batch_size),epochs=epochs,batch_size=batch_size,validation_data=test_dataset.shuffle(len(x_test)).batch(batch_size))
```

```python
model.save_pretrained("/tmp/sentiment_model")
```

```python
unseen_data=data.sample(frac=0.5,random_state=42)
unseen_review=unseen_data['review'].to_list()

unseen_encodings=tokenizer(unseen_review,padding=True,truncation=True,max_length=256,return_tensors="tf")
```

```python
y_unseen=unseen_data['sentiment'].tolist()
unseen_encodings=tf.data.Dataset.from_tensor_slices((dict(unseen_encodings),y_unseen))
```

```python
model.evaluate(unseen_encodings.shuffle(len(unseen_review)).batch(batch_size),return_dict=True,batch_size=batch_size)
```

```python
def predict(sentence):
  loaded_model=TFDistilBertForSequenceClassification.from_pretrained("tmp/sentiment_model")
  predict_input=tokenizer.encode(sentence,truncation=True,padding=True,return_tensor='tf')
  tf_output=loaded_model.predict(predict_input)[0]
  tf_prediction=tf.nn.softmax(tf_output,axis=1).numpy[0]
  return ['negative','positive'][np.argmax(tf_prediction)]
```

```python
demo=gr.Interface(fn=predict,inputs='text',outputs='text',live=True)
demo.launch()
```
